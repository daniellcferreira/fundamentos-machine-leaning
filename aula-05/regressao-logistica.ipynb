```python
def sigmoid(z):

    return 1/(1 + np.exp(-z))
```

```python
import ipywidgets

def plot_sigmoid(b0, b1):

    plt.title("Curva sigmoide: $y(x) = 1(1 + e^{-(b_0 + b_1 x)})$")

    x = np.linspace(-10, 10, 1000)

    y = sigmoid(b0 + b1*x)

    plt.plot(x, y, label=f"b0={b0:.2f} | b1={b1:.2f}")
    plt.legend()

    plt.ylim(-0.1, 1.1)

    plt.xlabel("x")
    plt.ylabel("y")

ipywidgets.interact(plot_sigmoid, b0=(-10, 10, 1), b1=(-10, 10, 1))
```
